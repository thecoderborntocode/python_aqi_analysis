```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
data = pd.read_csv('city_day.csv')

# Assuming 'City' and 'NO' are column names in your CSV file
# Replace 'City' and 'NO' with the actual column names from your DataFrame
plt.bar(data['City'], data['NO'])
plt.xlabel('City')
plt.ylabel('NO Levels')
plt.title('NO Levels in Different Cities')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
